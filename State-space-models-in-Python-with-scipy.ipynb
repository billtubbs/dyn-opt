# State-space models in Python

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
import sys, scipy, numpy; print(scipy.__version__, numpy.__version__, sys.version_info)
```

## SISO system

```python
a = -4
b = 2
c = 1
d = 0
sys = signal.StateSpace(a, b, c, d)
print(sys)
t, y = signal.step(sys)
print(y[0:5])
```

```python
u = np.ones_like(t)
t, y, x = signal.lsim(sys,u,t)
print(y[0:5])
```

```python
plt.plot(t, y)
plt.grid()
plt.show()
```

```python
t, y = signal.impulse(sys)
print(y[0:5])
```

```python
a = -4.0  # Type float
b = 2
c = 1
d = 0
sys = signal.StateSpace(a, b, c, d)
print(sys)
t, y = signal.step(sys)
print(y[0:5])
```

```python
plt.plot(t, y)
plt.grid()
plt.show()
```

## MIMO system

```python
a = np.array([[0, 1], [0, 0]])
b = np.array([[0], [1]])
c = np.array([[1, 0]])
d = np.array([[0]])
sys = signal.StateSpace(a, b, c, d)
t, y = signal.step(sys)
print(y[0:5])
```

```python
a = np.array([[0.0, 1], [0, 0]])
b = np.array([[0], [1]])
c = np.array([[1, 0]])
d = np.array([[0]])
sys = signal.StateSpace(a, b, c, d)
t, y = signal.step(sys)
print(y[0:5])
```
